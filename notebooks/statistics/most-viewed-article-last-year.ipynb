## Most viewed articles over last year

```python
from datetime import datetime, timedelta


def analyze(df):
    one_year_ago = datetime.now() - timedelta(days=365)
    df["page_name"] = df["page_name"].fillna("Mathe für Nicht-Freaks")
    df = df[(df["timestamp"] > one_year_ago)]
    df = df[["page_name", "section_name", "book_name", "views"]].groupby(["page_name", "section_name", "book_name"]).sum().reset_index()
    df.sort_values("views", ascending=False, inplace=True)

    return df


analyze(df.copy())
```
